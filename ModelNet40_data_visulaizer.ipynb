```python
import numpy as np
import open3d as o3d
```

```python
def downsample(points, num_points=1800):
    if len(points) > num_points:
        choice = np.random.choice(len(points), num_points, replace=False)
    points = points[choice, :]
    return points
```

```python
path = r"folder path\plant\train\plant_0015.npy"
points = np.load(path)

points = downsample(points, num_points=6000)

pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points)
# Visualize the point cloud
o3d.visualization.draw_geometries([pcd])
```

```python
o3d.visualization.draw_plotly([pcd])
```

```python
# .off file visulaization/ mesh visualization

import open3d as o3d
import numpy as np

# Load the .off file
path = r"folder path \plant\train\plant_0007.off"
mesh = o3d.io.read_triangle_mesh(path)

# Compute vertex normals
mesh.compute_vertex_normals()

# Assign colors based on vertex normals
colors = np.abs(np.array(mesh.vertex_normals))
colors = colors / np.max(colors)
mesh.vertex_colors = o3d.utility.Vector3dVector(colors)

# Visualize the colored mesh
o3d.visualization.draw_geometries([mesh])
```
